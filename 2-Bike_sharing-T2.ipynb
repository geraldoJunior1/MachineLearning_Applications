```python
df_media_season = df_bike.groupby(['season']).total_count.mean().reset_index(name='media')
display(df_media_season)


print(f'Maior média é a season {df_media_season[df_media_season.media==df_media_season.media.max()].season.iloc[0]}') 
print(f'Menor média é a season {df_media_season[df_media_season.media==df_media_season.media.min()].season.iloc[0]}')
```

```python
df_media_hour = df_bike.groupby(['hour']).total_count.mean().reset_index(name='media')
display(df_media_hour)


print(f'Maior média é a hora {df_media_hour[df_media_hour.media==df_media_hour.media.max()].hour.iloc[0]}') 
print(f'Menor média é a hora {df_media_hour[df_media_hour.media==df_media_hour.media.min()].hour.iloc[0]}')
```

```python
df_media_weekday= df_bike.groupby(['weekday']).total_count.mean().reset_index(name='media')
display(df_media_weekday)


print(f'Maior média é a hora {df_media_weekday[df_media_weekday.media==df_media_weekday.media.max()].weekday.iloc[0]}') 
print(f'Menor média é a hora {df_media_weekday[df_media_weekday.media==df_media_weekday.media.min()].weekday.iloc[0]}')
```

```python
df_media_hora = df_bike[df_bike.weekday==3].groupby(['hour']).total_count.mean().reset_index(name='media')
display(df_media_hora)

print(f'Maior média é as {df_media_hora[df_media_hora.media==df_media_hora.media.max()].hour.iloc[0]} horas')
```

```python
df_media_hora = df_bike[df_bike.weekday==6].groupby(['hour']).total_count.mean().reset_index(name='media')
display(df_media_hora)

print(f'Maior média é as {df_media_hora[df_media_hora.media==df_media_hora.media.max()].hour.iloc[0]} horas')
```
