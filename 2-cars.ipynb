```python
df_car = pd.read_csv('imports-85.txt', sep=",", header=None,na_values="?")
df_car.info()
```

```python
titulos = {0:"symboling", 1:"normalized_losses", 2:"make", 3:"fuel_type", 4:"aspiration",
           5:"num_doors", 6:"body_style", 7:"drive_wheels",8: "engine_location",
           9:"wheel_base", 10:"length", 11:"width",12: "height", 13:"curb_weight",
           14:"engine_type", 15:"num_cylinders",16: "engine_size", 17:"fuel_system",
           18:"bore", 19:"stroke", 20:"compression_ratio", 21:"horsepower", 22:"peak_rpm",
           23:"city_mpg", 24:"highway_mpg", 25:"price"}

df_car.rename(columns=titulos,inplace=True)
df_car.info()
```

```python
df_car_objects = df_car.select_dtypes(include=['object']).copy()

df_car_objects.info()
```

```python
df_car_objects.isnull().sum()
```

```python
df_car_objects.dropna(inplace=True)
df_car_objects.isnull().sum()
```

```python
df_car_objects["num_cylinders"].value_counts()
```

```python
map_classes = {"num_doors":     {"four": 4, "two": 2},
                "num_cylinders": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}
```

```python
df_car_objects.replace(map_classes, inplace=True)
df_car_objects.info()
```

```python
df_car_objects["body_style"] = df_car_objects["body_style"].astype('category')
df_car_objects.info()
```
