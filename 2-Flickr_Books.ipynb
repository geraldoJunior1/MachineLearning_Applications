```python
df_books.loc[1905:, 'Date of Publication'].head(15) # a partir da linha 1905 pra frente na coluna data
```

```python
regular = r'^(\d{4})'  #expressao regular
extr = df_books['Date of Publication'].str.extract(regular, expand=False) #arrumando str para a expressao

extr[1905:1920]
```

```python
df_books['Date of Publication']=pd.to_numeric(extr) #transforma pra numerico
```

```python
df_books['Date of Publication'].isnull().sum() #quantos nulos
```

```python
df_books.dropna(subset=['Date of Publication'],inplace=True) #tira os nulos
df_books['Date of Publication']=df_books['Date of Publication'].astype(int)  #transforma float em int
```

```python
df_books['Place of Publication'].head(15)
```

```python
#coloca certinho os nomes das cidades e tira os bairros
london = df_books['Place of Publication'].str.contains('London')  
oxford = df_books['Place of Publication'].str.contains('Oxford')

df_books['Place of Publication']=np.where(london, 'London',np.where(oxford, 'Oxford',df_books['Place of Publication'].str.replace('-', ' ')))
df_books['Place of Publication'].head(15)
```
